```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

data = pd.read_csv('epi_r.csv')
data
```

```python
data_1 = data[['title','rating','calories','protein','fat','sodium','22-minute meals','3-ingredient recipes','dessert']]
data_1
```

# DATA CLEANING

```python
data_1.columns = data_1.columns.str.replace('-','_',regex=True)
data_1.columns = data_1.columns.str.replace(' ','_',regex=True)
```

```python
data_1
```

```python
data_1.info()
```

```python
data_1.describe()
```

```python
data_1.isnull().sum()
```

```python
print('maximum calories : \n',data_1['calories'].max())
print('minimum calories : \n',data_1['calories'].min())
```

```python
print('maximum proteins : \n',data_1['protein'].max())
print('minimum proteins : \n',data_1['protein'].min())
```

```python
print('maximum fat : \n',data_1['fat'].max())
print('minimum fat : \n',data_1['fat'].min())
```

```python
print('maximum sodium : \n',data_1['sodium'].max())
print('minimum sodium : \n',data_1['sodium'].min())
```

```python
cal = data_1['calories'].mean()
pro = data_1['protein'].mean()
fat = data_1['fat'].mean()
sod = data_1['sodium'].mean()
print('mean of the calories      :  ',cal)
print('mean of the proteins      :  ',pro)
print('mean of the fat           :  ',fat)
print('mean of the sodium        :  ',sod)
```

```python
data_1['calories'].fillna(cal,inplace=True)
#data_1['protein'].fillna(pro,inplace=True)
data_1['fat'].fillna(fat,inplace=True)
data_1['sodium'].fillna(sod,inplace=True)
data_1.isnull().sum()
```

```python
data_1.rename(columns={'protien': 'protein'}, inplace=True)
data_1['protein'].fillna(pro,inplace=True)
data_1.isnull().sum()
```

```python
data_1[data_1.duplicated()]
```

```python
data_c = data_1.drop_duplicates()
data_c = data_c.reset_index()
data_c
```

```python
data_c = data_c.drop(columns=['index'])
data_c
```

```python
data_c.rating.value_counts()
```

```python
replace = {
    4.375:4,3.750:4,3.125:3,
    2.500:3,1.250:1,1.875:2
}
data_c['rating'] = data_c['rating'].replace(replace)
data_c['rating'] = data_c['rating'].astype(int)
data_c.rating.value_counts()
```

```python
data_c['22_minute_meals'] = data_c['22_minute_meals'].astype(int)
data_c['3_ingredient_recipes'] = data_c['3_ingredient_recipes'].astype(int)
data_c['dessert'] = data_c['dessert'].astype(int)
```

```python
data_c.to_csv('epi_r_clean_data.csv',index=False)
```

# 3.How can the data help improve the user experience for a recipe platform?

#### Provide recommendations based on balanced nutritional content (e.g., moderate calories, fat, and sodium).

```python
# Function to recommend balanced diet recipes.
def balanced_recipes():
    balanced_recipes = data_c[(data_c['calories'] <= 500) & 
                            (data_c['fat'] <= 20) & 
                            (data_c['sodium'] <= 300)]
    
    return balanced_recipes[['title', 'rating', 'calories', 'protein', 'fat', 'sodium']].head(10)

balanced_recipes()
```
